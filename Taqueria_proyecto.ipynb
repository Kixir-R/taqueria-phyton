# programa taqueria

en este programa vamos a crear un sistema de software que ayude a una taquería a controlar sus productos y almecenar sus datos..
 
 usaremos archivos JSON y codigo phyton para guardar los datos y no perder los datos cuando se cierre el programa.

```python
import json

productos = []
```

## Guardar datos en archivo JSON

Esta función se encarga de guardar todos los productos en un archivo llamado productos.json

Asi, aunque cerremos el programa, la información se conserva

```python
def guardar_datos():
    archivo = open("productos.json", "w")
    json.dump(productos, archivo, ensure_ascii=False, indent=4)
    archivo.close()
```

## Cargar datos

Esta función intenta abrir el archivo de productos.json

Si el archivo existe, carga los datos a la lista productos.
Si no, simplemente inicia una lista vacía para evitar errores.

```python
def cargar_datos():
    global productos
    try:
        archivo = open("productos.json", "r")
        productos = json.load(archivo)
        archivo.close()
    except:
        productos = []
        
```

## <font color= "#ffea00"> **Definicion para dar de alta los productos**

#### <font color= "#ffffff"> Mediante la función input, se definirá los distintos productos que se darán de alta, como el nombre, código, precio y la cantidad, en caso de no agregar un dato, se le indicará al usuario que no puede estar vacío y se le volverá a solicitar el dato. Una vez que se hayan ingresado todos los datos, se creará un diccionario con la información del producto y se agregará a la lista de productos, para luego guardar los datos en el archivo JSON.

```python
def agregar_producto():
    print("\nALTA DE PRODUCTO")

    nombre = input("Nombre: ")
    while nombre == "":
        print("No puede estar vacío")
        nombre = input("Nombre: ")

    codigo = input("Código: ")
    while codigo == "":
        print("No puede estar vacío")
        codigo = input("Código: ")

    while True:
        precio = input("Precio: ")
        try:
            precio = float(precio)
            if precio >= 0:
                break
            else:
                print("Debe ser positivo")
        except:
            print("Ingresa un número válido")

    while True:
        cantidad = input("Cantidad: ")
        try:
            cantidad = int(cantidad)
            if cantidad >= 0:
                break
            else:
                print("Debe ser positivo")
        except:
            print("Ingresa un número entero válido")

    producto = {
        "nombre": nombre,
        "codigo": codigo,
        "precio": precio,
        "cantidad": cantidad
    }

    productos.append(producto)
    guardar_datos()

    print("Producto guardado correctamente\n")
```

## <font color= "#ffea00"> **Definicion para mostrar la lista de productos**

#### <font color= "#ffffff"> Esta función muestra la lista de productos registrados en el sistema. Si no hay productos, se muestra un mensaje indicando que no hay productos registrados. Si hay productos, se muestra una tabla con el nombre, código, precio y cantidad de cada producto.

```python
def mostrar_productos():
    print("\nLISTA DE PRODUCTOS")

    if len(productos) == 0:
        print("No hay productos")
    else:
        for p in productos:
            print("Nombre:", p["nombre"], "| Código:", p["codigo"], "| Precio:", p["precio"], "| Cantidad:", p["cantidad"])
    print()

cargar_datos()

opcion = ""
while opcion != "3":
    print("1. Agregar producto")
    print("2. Mostrar productos")
    print("3. Salir")

    opcion = input("Elige una opción: ")

    if opcion == "1":
        agregar_producto()
    elif opcion == "2":
        mostrar_productos()
    elif opcion == "3":
        print("Adiós")
    else:
        print("Opción incorrecta\n")
```

## Mostrar productos

Esta función imprime en pantalla todos los productos disponibles.

Muestra el nombre, precio y cantidad de cada producto en forma de lista para que sea fácil de leer.

```python
guardar_datos()
cargar_datos()
print(productos)
```
