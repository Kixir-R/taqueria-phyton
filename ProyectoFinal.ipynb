## Menu para edicion de Menu
#### Usaremos funciones para crear, editar, borrar, buscar, etc. en el archivo json

```python
import json
productos = []
menu = 0

##Guardar datos en archivo JSON
# Esta función se encarga de guardar todos los productos en un archivo llamado productos.json
# Asi, aunque cerremos el programa, la información se conserva

def guardar_datos():
        archivo = open("productos.json", "w")
        json.dump(productos, archivo, ensure_ascii=False, indent=4)
        archivo.close()

##Esta función intenta abrir el archivo de productos.json
# Si el archivo existe, carga los datos a la lista productos.
# Si no, simplemente inicia una lista vacía para evitar errores.

def cargar_datos():
        global productos
        try:
            archivo = open("productos.json", "r")
            productos = json.load(archivo)
            archivo.close()
        except:
            productos = []

def editar_menu():

    ##  Mediante la función input, se definirá los distintos productos que se darán de alta,
    #  como el nombre, código, precio y la cantidad, en caso de no agregar un dato,
    #  se le indicará al usuario que no puede estar vacío y se le volverá a solicitar el dato. 
    # Una vez que se hayan ingresado todos los datos, 
    # se creará un diccionario con la información del producto y se agregará a la lista de productos, 
    # para luego guardar los datos en el archivo JSON.

    def agregar_producto():
        print("\nALTA DE PRODUCTO")

        nombre = input("Nombre: ")
        while nombre == "":
            print("No puede estar vacío")
            nombre = input("Nombre: ")

        codigo = input("Código: ")
        while codigo == "":
            print("No puede estar vacío")
            codigo = input("Código: ")

        while True:
            precio = input("Precio: ")
            try:
                precio = float(precio)
                if precio >= 0:
                    break
                else:
                    print("Debe ser positivo")
            except:
                print("Ingresa un número válido")

        while True:
            cantidad = input("Cantidad: ")
            try:
                cantidad = int(cantidad)
                if cantidad >= 0:
                    break
                else:
                    print("Debe ser positivo")
            except:
                print("Ingresa un número entero válido")

        producto = {
            "nombre": nombre,
            "codigo": codigo,
            "precio": precio,
            "cantidad": cantidad
        }

        productos.append(producto)
        guardar_datos()
        print("Producto guardado correctamente\n")

    ##Aqui se define las variables para buscar los productos eliminarlos y 
    # editarlos segun como el ususario quiera,
    #  si  el usuario ingresa un productio no existente este programa le dira que dicho producto
    #  no existe en el programa.

    def buscar_producto():
        print("\nBUSCAR PRODUCTO")
        criterio = input("Ingresa nombre o código: ").lower()

        encontrados = []
        for p in productos:
            if criterio in p["nombre"].lower() or criterio in p["codigo"].lower():
                encontrados.append(p)

        if len(encontrados) == 0:
            print("No se encontraron productos\n")
        else:
            for p in encontrados:
                print("Nombre:", p["nombre"], "| Código:", p["codigo"], "| Precio:", p["precio"], "| Cantidad:", p["cantidad"])
        print()


    def eliminar_producto():
        print("\nELIMINAR PRODUCTO")
        codigo = input("Ingresa el código del producto a eliminar: ")

        for p in productos:
            if p["codigo"] == codigo:
                productos.remove(p)
                guardar_datos()
                print("Producto eliminado correctamente\n")
                return

        print("Producto no encontrado\n")


    def editar_producto():
        print("\nEDITAR PRODUCTO")
        codigo = input("Ingresa el código del producto a editar: ")

        for p in productos:
            if p["codigo"] == codigo:
                print("Deja vacío si no quieres cambiar el valor")

                nuevo_nombre = input("Nuevo nombre: ")
                if nuevo_nombre != "":
                    p["nombre"] = nuevo_nombre

                nuevo_precio = input("Nuevo precio: ")
                if nuevo_precio != "":
                    try:
                        p["precio"] = float(nuevo_precio)
                    except:
                        print("Precio inválido, se mantiene el anterior")

                nueva_cantidad = input("Nueva cantidad: ")
                if nueva_cantidad != "":
                    try:
                        p["cantidad"] = int(nueva_cantidad)
                    except:
                        print("Cantidad inválida, se mantiene la anterior")

                guardar_datos()
                print("Producto actualizado correctamente\n")
                return

        print("Producto no encontrado\n")
        
    ## Esta función muestra la lista de productos registrados en el sistema.
    #Si no hay productos, se muestra un mensaje indicando que no hay productos registrados. 
    # Si hay productos, se muestra una tabla con el nombre, código, precio y cantidad de cada producto. 
    # Igualmente mostrara si quiere eliminar o editar uno de los productos.

    def mostrar_productos():
        print("\nLISTA DE PRODUCTOS")

        if len(productos) == 0:
            print("No hay productos")
        else:
            for p in productos:
                print("Nombre:", p["nombre"], "| Código:", p["codigo"], "| Precio:", p["precio"], "| Cantidad:", p["cantidad"])
        print()

    cargar_datos()


    opcion = ""
    while opcion != "6":
        print("1. Agregar producto")
        print("2. Mostrar productos")
        print("3. Buscar producto")
        print("4. Editar producto")
        print("5. Eliminar producto")
        print("6. Salir")

        opcion = input("Elige una opción: ")

        if opcion == "1":
            agregar_producto()
        elif opcion == "2":
            mostrar_productos()
        elif opcion == "3":
            buscar_producto()
        elif opcion == "4":
            editar_producto()
        elif opcion == "5":
            eliminar_producto()
        elif opcion == "6":
            print("Adiós")
        else:
            print("Opción incorrecta\n")
```

## Menu de compra
#### Usaremos una funcion para poder crear el menu y submenus de compra, modificar el archivo json y crear el carrito y total a pagar

```python
def compra_venta():
    ## Cargamos el archivo con permisos de lectura para mostrar el menu,
    #  solo en lectura para no hechar a perder lo ya hecho

    with open("productos.json", "r", encoding="utf-8") as archivo:
        productos = json.load(archivo)

    ##se necesita de otra lista para guardar el carrito para comprar 

    carrito = []

    ##Usamos un ciclo for recorrer la lista, e imprimimos nombre, producto y precio de los productoas
    #ademas de usar un contador para enumerarlos

    def mostrar_menu():
        print("MENÚ \n")
        for i, producto in enumerate(productos):
            print(f"{i+1}. {producto['nombre']} - precio: {producto['precio']} (Stock: {producto['cantidad']})")
    
    def comprar_producto():
        
        ##Usamos un try para prevenir errores al final de la funcion

        mostrar_menu()
        
        try:
            opcion = int(input("\nSelecciona el número del producto que deseas comprar: ")) - 1
            
            ##evitamos que se pidan productos inexistentes o que no se enumeren correctamente

            if opcion < 0 or opcion >= len(productos):
                print("Opción inválida.")
                return
            
            ##Guardamos el producto seleccionado y pedimos la cantidad para sacar el precio despues
            producto = productos[opcion]
            
            cantidad = int(input(f"¿Cuántos '{producto['nombre']}' deseas comprar?: "))
            
            ##Comprobamos que no se ordenen mas o menos de la cantidad disponible

            if cantidad <= 0:
                print("Cantidad inválida.")
                return
            
            if cantidad > producto['cantidad']:
                print("No hay suficiente stock.")
                return
            
            # Actualizamos el stock quitandole la cantidad a comprar
            producto['cantidad'] -= cantidad
            
            # modificamos momentanemente los permisos a escritura para modificar el stock
            with open("productos.json", "w", encoding="utf-8") as archivo:
                json.dump(productos, archivo, indent=4, ensure_ascii=False)
            
            # agregamos a la lista de carrito el nombre, presio de el producto selecccionado

            carrito.append({
                "nombre": producto["nombre"],
                "cantidad": cantidad,
                "subtotal": cantidad * producto["precio"]
            })
            
            print("Producto agregado al carrito.")

        ## evitamos errores aqui con execpt

        except ValueError:
            print("Entrada inválida.")

    def mostrar_carrito():
        print("CARRITO DE COMPRAS\n")
        
        total = 0
        
        ##Recorremos lista de carrito para obtener el total
        for item in carrito:
            print(f"{item['nombre']} x{item['cantidad']} = ${item['subtotal']}")
            total += item["subtotal"]
        
        print(f"TOTAL: ${total}")
        return total

    ##Imprimimos el menu de compra como funcion para llamarla y solo imprimirla una vez
    # cada que se seleccione en el menu principal
    
    def menu_compra():
        
        while True:
            print("\n1. Comprar producto")
            print("2. Ver carrito")
            print("3. Salir")
            
            opcion = input("Selecciona una opcion: ")
            
            if opcion == "1":
                comprar_producto()
            elif opcion == "2":
                mostrar_carrito()
            elif opcion == "3":
                print("Gracias por tu compra.")
                break
            else:
                print("Opción inválida.")

    ##Llamamos inmediatamenta la funcion de menu fuera de la funcion para imprimir el menu de compra

    menu_compra()
```

## Menu principal
#### a traves de este menu accederemos a el menu de compra y al menu para modificar el menu diponible, usamos funciones para crear menus dentro de menus

```python
while menu != 3:
    print("Menu\n")

    try:
        print("1. editar menu disponible")
        print("2. menu de compra")
        print("3. salir\n")
        menu = int(input("Ingrese la opcion a elegir"))
        if menu == 1:
            editar_menu()
        elif menu == 2:
            compra_venta()
        elif menu == 3:
            print("adios")
        else:
            print("opcion invalida")

    except ValueError:
        print("Por favor, ingresa un número válido.")

print("gracias por usar nustro software")
menu = 0
```
